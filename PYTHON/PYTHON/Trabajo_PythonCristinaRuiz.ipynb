#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
# Diccionario con las enzimas de restricción y sus secuencias de reconocimiento.
# El símbolo "*" indica el punto exacto donde la enzima corta la cadena de ADN.
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI'   : 'G*GATCC',
    'AluI'    : 'AG*CT',
    'Sau3AI'  : '*GATC',
    'EcoRI'   : 'G*AATTC'
}

# Función principal que simula la digestión de ADN con enzimas de restricción.
# Busca los sitios de corte en la secuencia (sec2), genera los fragmentos
# y calcula su número y tamaño.
def digerir():

    # Creamos tres estructuras:
    # - Un diccionario para guardar los cortes de cada enzima.
    # - Una lista para las enzimas que no cortan.
    # - Una lista con todas las posiciones de corte encontradas.
    cortes_por_enz = {}
    sin_sitio = []
    todos_cortes = []

    # Recorremos el diccionario con las enzimas y sus secuencias de reconocimiento.
    for nombre, patron in enzR.items():

        # Buscamos la posición del símbolo "*" dentro del patrón.
        star_pos = patron.index("*")

        # Eliminamos el "*" del patrón para poder buscarlo en la secuencia.
        motivo = patron.replace("*", "")

        # Lista donde se guardarán las posiciones de corte de esta enzima.
        posiciones = []

        # Variable de inicio para buscar dentro de la secuencia.
        i = 0

        # Usamos un bucle que va buscando todas las coincidencias del motivo en la secuencia.
        while True:
            # find() devuelve la posición donde se encuentra el motivo o -1 si ya no hay más.
            idx = sec2.find(motivo, i)
            if idx == -1:
                break  # si no hay más coincidencias, salimos del bucle

            # Calculamos el punto de corte sumando la posición del símbolo "*".
            corte0 = idx + star_pos

            # Guardamos la posición de corte (+1 por convención biológica).
            posiciones.append(corte0 + 1)

            # También lo añadimos a la lista general de cortes.
            todos_cortes.append(corte0)

            # Avanzamos una posición para seguir buscando.
            i = idx + 1

        # Si la enzima tiene sitios de corte, los guardamos en el diccionario.
        if posiciones:
            cortes_por_enz[nombre] = posiciones
        # Si no tiene sitios, la añadimos a la lista sin_sitio.
        else:
            sin_sitio.append(nombre)

    # Una vez encontrados todos los cortes, eliminamos duplicados con set()
    # y los ordenamos de menor a mayor con sorted().
    todos_cortes = sorted(set(todos_cortes))

    # Lista donde se guardarán los fragmentos generados tras los cortes.
    fragmentos = []

    # Variable para marcar el inicio de cada fragmento.
    prev = 0

    # Recorremos todas las posiciones de corte para extraer los fragmentos.
    for c in todos_cortes:
        # Guardamos el fragmento desde el corte anterior hasta el actual.
        fragmentos.append(sec2[prev:c])
        # Actualizamos el inicio del siguiente fragmento.
        prev = c

    # Agregamos el último fragmento, que va desde el último corte hasta el final de la secuencia.
    fragmentos.append(sec2[prev:])

    # Calculamos el tamaño de cada fragmento midiendo su longitud en bases.
    tamaños = [len(f) for f in fragmentos]

    # Devolvemos toda la información: 
    # cortes por enzima, enzimas sin sitio, los fragmentos y sus tamaños.
    return cortes_por_enz, sin_sitio, fragmentos, tamaños



# Ejecutamos la digestión y guardamos los resultados en variables separadas.
cortes, no_sites, frags, sizes = digerir()

# Mostramos la información obtenida.
print("Cortes por enzima:", cortes)
print("Enzimas sin sitio:", no_sites)
print("Fragmentos generados:", len(frags))
print("Tamaños de fragmentos:", sizes)
```

```python
#Este print solo dibuja un encabezado en la salida para separar visualmente el informe.
print("=== Resultado de la digestión ===")

if cortes:
    # Muestra el número total de enzimas que sí tienen sitios de corte.
    # len(cortes) devuelve cuántas claves tiene el diccionario.
    print(f"Enzimas que cortan ({len(cortes)}):")
    # Recorremos el diccionario: 'enz' es el nombre de la enzima, 'pos' la lista de posiciones.
    for enz, pos in cortes.items():
        # 'sorted(pos)' ordena las posiciones de corte de menor a mayor para que sea más legible.
        # '{len(pos)}' indica cuántos cortes realiza esa enzima en la secuencia.
        print(f"  - {enz}: {len(pos)} cortes en posiciones {sorted(pos)}")
 # Si ninguna enzima encontró su sitio de reconocimiento, lo indicamos.
else:
    print("Ninguna enzima encontró sitio de corte.")

# Enzimas sin sitio
if no_sites:
    # Mostramos cuántas y cuáles son, unidas por comas para mejor lectura.
    print(f"Enzimas sin sitio ({len(no_sites)}): {', '.join(no_sites)}")

# Fragmentos resultantes con tamaño y coordenadas dentro de la secuencia original (0-based)
print(f"\nFragmentos resultantes: {len(frags)}")
# Vamos a imprimir también las coordenadas de cada fragmento en la secuencia original
# usando notación Python (0-based, [inicio:fin) ), para que sea clara la segmentación.
inicio = 0 
# 'inicio' guarda el índice 0-based donde empieza el fragmento actual en la secuencia original.
# 'zip(frags, sizes)' empareja cada fragmento con su tamaño.
# 'enumerate(..., start=1)' numera los fragmentos comenzando en 1 para presentación humana.
for i, (frag, tam) in enumerate(zip(frags, sizes), start=1):
    # El final (exclusivo) del fragmento en notación Python es inicio + tamaño.
    fin = inicio + tam
    # Construimos una vista previa (preview) de la secuencia del fragmento:
    # - Si el fragmento es largo (>20 bp), mostramos los 10 primeros y 10 últimos nucleótidos con "..." en medio.
    # - Si es corto (≤20 bp), mostramos el fragmento completo.
    preview = f"{frag[:10]}...{frag[-10:]}" if tam > 20 else frag
     # Imprimimos una línea descriptiva por fragmento con:
    #   - número de fragmento (i)
    #   - tamaño en pares de bases (tam)
    #   - coordenadas 0-based [inicio:fin] en la secuencia original
    #   - la vista previa de la secuencia
    print(f"  Fragmento {i}: {tam} bp  [{inicio}:{fin}]  {preview}")
    # Actualizamos 'inicio' para que el siguiente fragmento empiece donde terminó el actual.
    # Esto asume que 'frags' y 'sizes' representan una partición consecutiva de la secuencia original
    inicio = fin

```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
def encontrar_orfs(secuencia):  # define una función que recibe una secuencia (parámetro de entrada)
    # Conjunto de codones STOP que terminan un ORF
    stop_codones = {"TAA", "TAG", "TGA"}
    # Aquí iremos acumulando los ORFs encontrados en ambos sentidos, 
    # lista vacía donde iremos guardando resultados
    orfs = []
    # Longitud de la secuencia
    n = len(secuencia)
    # diccionario para obtener la base complementaria
    comp = {"A": "T", "T": "A", "C": "G", "G": "C"}

    # Hebra positiva: buscamos ORFs (de ATG a STOP) leyendo la cadena  (5'→3')
     # probamos los 3 marcos de lectura posibles: 0, 1 y 2
    for frame in range(3): 
        # i empieza alineado al frame (0, 1 o 2)
        i = frame   
        # mientras quede al menos un codón completo (3 nt) por leer
        while i <= n - 3:           
             # extraemos el codón actual: posiciones i, i+1, i+2
            codon = secuencia[i:i+3]  
             # si encontramos el codón de inicio (Met)
            if codon == "ATG":       
                # empezamos a buscar un STOP justo en el siguiente codón (mismo frame)
                j = i + 3             
                # seguimos avanzando codón a codón dentro del mismo frame
                while j <= n - 3:       
                    codon_stop = secuencia[j:j+3]   # leemos el codón en la posición j (3 nt)
                    if codon_stop in stop_codones:  # si encontramos un STOP válido (TAA, TAG o TGA)
                        orf = secuencia[i:j+3]      # ORF incluye desde el ATG incluido hasta el STOP incluido
                        # guardamos la info del ORF encontrado:
                        # "positive" significa hebra positiva; frame → 0/1/2; i → inicio 0-based;
                        # j+3 → final exclusivo (estilo slicing de Python); orf → secuencia nucleotídica del ORF
                        orfs.append(("positive", frame, i, j+3, orf))
                        break
                    j += 3                 # saltamos al siguiente codón (manteniendo el frame) para seguir buscando STOP
            i += 3                         # saltamos al siguiente codón dentro del mismo frame para seguir escaneando

    # Hebra negativa (reverse-complement)
    for frame in range(3):                     # también hay 3 marcos en la hebra opuesta: 0, 1 y 2
        i = n - frame - 3                      # punto de arranque alineado al frame desde el final (para poder leer i:i+3)
        while i >= 0:                          # recorremos hacia atrás (porque la hebra complementaria se lee en sentido inverso)
            # "".join(...) une caracteres en una cadena nueva
            # secuencia[i:i+3] tomamos 3 nt en la hebra original
            # [::-1] los invertimos (antiparalelismo 3'→5' → 5'→3')
            # comp[b] sustituimos cada base por su complementaria (A↔T, C↔G)
            # Resultado: el codón tal y como se vería en la reverse-complement (5'→3')
            codon = "".join(comp[b] for b in secuencia[i:i+3][::-1])
            if codon == "ATG":                 # si en la reverse-complement vemos un ATG → inicio de ORF en hebra negativa
                j = i                          # empezamos a retroceder de 3 en 3 para buscar el STOP en el mismo frame
                while j >= 0:
                    # mismo proceso para el codón de parada pero en la posición j
                    stop_codon = "".join(comp[b] for b in secuencia[j:j+3][::-1])
                    if stop_codon in stop_codones:         # STOP válido (TAA/TAG/TGA) en la reverse-complement
                        orf = secuencia[j:i+3]             # tramo en la cadena original que abarca desde STOP hasta ATG (incluido)
                        # guardamos como
                        # "Negativa" referente a lahebra,
                        # el frame 0/1/2,
                        # j+1 inicio reportado en numeración biológica (1-based),
                        # i+3 final exclusivo (estilo slicing 0-based),
                        # orf  secuencia nucleotídica del ORF en la hebra original
                        orfs.append(("Negativa", frame, j+1, i+3, orf))  # append() añade esta tupla al listado
                        break
                    j -= 3                      # retrocedemos un codón (mantenemos el frame en la hebra negativa)
            i -= 3                              # seguimos retrocediendo codón a codón en este frame

    return orfs   # devolvemos la lista con todos los ORFs encontrados (positiva + negativa)


# encontrar_orfs(sec2) en "sec2" 
for orf in encontrar_orfs(sec2):
    print(orf)    # print()  muestra por pantalla cada tupla ORF (hebra, frame, inicio, fin, secuencia)
```

```python
def seis_mas_largos(orfs):   # define una función que recibe una lista de ORFs (tuplas con info de cada marco abierto de lectura)


    if not orfs:   # comprobamos si la lista está vacía o None (no hay ORFs)
        print("No se encontraron ORFs.")   # mensaje de aviso si no se encontraron
        return []   # salimos de la función y devolvemos una lista vacía

    # Ordenamos los ORFs por su longitud (de mayor a menor)
    # sorted() crea una nueva lista ordenada
    # key=lambda x: len(x[4]) → usamos como criterio de orden la longitud de la secuencia ORF (posición 4 de la tupla)
    # reverse=True → para ordenar de mayor a menor
    orfs_ordenados = sorted(orfs, key=lambda x: len(x[4]), reverse=True)

    # Seleccionamos los 6 primeros elementos de la lista ya ordenada (los más largos)
    top6 = orfs_ordenados[:6]

    # Imprimimos un encabezado para mostrar los resultados
    print("\n=== Los 6 ORFs más largos ===")

    # Recorremos la lista top6 usando enumerate() para numerar del 1 al 6 automáticamente
    # Cada elemento se desempaqueta en hebra, frame, start, end, seq
    for i, (hebra, frame, start, end, seq) in enumerate(top6, 1):

        # Creamos una variable 'strand' con el símbolo + o - según la hebra
        # Si la palabra 'hebra' empieza por "pos" (positive), usamos "+"
        # Si no, usamos "-"
        strand = "+" if str(hebra).lower().startswith("pos") else "-"

        # Mostramos por pantalla toda la información de cada ORF
        # f-string → permite incluir variables dentro del texto fácilmente
        # len(seq) → longitud de la secuencia (número de nucleótidos del ORF)
        print(f"{i}. Strand: {strand} ({hebra}) | Frame: {frame} | Posición: {start}-{end} | Longitud: {len(seq)} nt")

    # Devolvemos la lista de los 6 ORFs más largos (por si queremos usarla después)
    return top6


#  Ejecución del flujo principal

# Primero llamamos a la función encontrar_orfs() para detectar todos los ORFs en la secuencia sec2
orfs = encontrar_orfs(sec2)   

# Luego pasamos esa lista a seis_mas_largos() para obtener e imprimir los 6 más largos
top6 = seis_mas_largos(orfs)
```

```python
def mejor_orf(orfs):   # define una función que recibe una lista de ORFs y busca el más largo (la mejor predicción)
    
    # Si la lista de ORFs está vacía o None, se muestra un mensaje y se detiene la función
    if not orfs:
        print("No se encontraron ORFs.")
        return None   # significa “sin resultado”

    # Encontrar el ORF más largo usando max()
    # max(iterable, key=...) → devuelve el elemento con el valor máximo según la clave indicada
    # Aquí usamos como clave len(x[4]) → la longitud de la secuencia (índice 4 de la tupla)
    best = max(orfs, key=lambda x: len(x[4]))

    # Desempaquetamos la tupla del ORF más largo en variables individuales
    # Esto mejora la legibilidad y permite acceder fácilmente a cada campo
    hebra, frame, start, end, seq = best

    # Creamos una variable 'strand' para mostrar el símbolo + o -
    # Si la hebra comienza por “pos” (positive), será +, si no, -
    strand = "+" if str(hebra).lower().startswith("pos") else "-"

    # Imprimimos los datos del ORF más largo con formato claro y organizado
    print("\n=== Mejor predicción (ORF más largo) ===")  # encabezado
    print(f"Strand: {strand} ({hebra})")                # hebra positiva o negativa
    print(f"Frame: {frame}")                            # marco de lectura
    print(f"Posición: {start}-{end}")                   # posiciones de inicio y fin
    print(f"Longitud: {len(seq)} nt")                   # tamaño total del ORF (nucleótidos)
    print(f"Secuencia del ORF:\n{seq}\n")               # muestra la secuencia completa del ORF

    # Devuelve el ORF más largo (la mejor predicción)
    return best


#  Ejecución del flujo completo 

# Llamamos a la función encontrar_orfs() para detectar todos los ORFs en la secuencia sec2
orfs = encontrar_orfs(sec2)   

# Llamamos a seis_mas_largos() para ver los 6 ORFs más largos y guardarlos
top6 = seis_mas_largos(orfs)

# Llamamos a mejor_orf() para determinar cuál de todos es el más largo (el mejor candidato a gen)
best = mejor_orf(orfs)
```

```python
# Tabla estándar codones de ADN que codifican para aminoácido 
# Diccionario que traduce cada triplete (codón) de ADN a su correspondiente aminoácido.
# "*" representa un codón STOP (fin de traducción).
CODON = {
    "TTT":"F","TTC":"F","TTA":"L","TTG":"L","CTT":"L","CTC":"L","CTA":"L","CTG":"L",
    "ATT":"I","ATC":"I","ATA":"I","ATG":"M","GTT":"V","GTC":"V","GTA":"V","GTG":"V",
    "TCT":"S","TCC":"S","TCA":"S","TCG":"S","AGT":"S","AGC":"S",
    "CCT":"P","CCC":"P","CCA":"P","CCG":"P",
    "ACT":"T","ACC":"T","ACA":"T","ACG":"T",
    "GCT":"A","GCC":"A","GCA":"A","GCG":"A",
    "TAT":"Y","TAC":"Y",
    "CAT":"H","CAC":"H","CAA":"Q","CAG":"Q",
    "AAT":"N","AAC":"N","AAA":"K","AAG":"K",
    "GAT":"D","GAC":"D","GAA":"E","GAG":"E",
    "TGT":"C","TGC":"C","TGG":"W",
    "CGT":"R","CGC":"R","CGA":"R","CGG":"R","AGA":"R","AGG":"R",
    "GGT":"G","GGC":"G","GGA":"G","GGG":"G",
    "TAA":"*","TAG":"*","TGA":"*"
}

STOP = {"TAA","TAG","TGA"}      # conjunto con los 3 codones STOP
AROM = {"F","Y","W"}            # conjunto con los aminoácidos aromáticos (Fenilalanina, Tirosina, Triptófano)


# para traducir ORF de ADN a proteína 
def traducir_orf(orf_nt: str) -> str:  # recibe la secuencia de ADN (orf_nt) y devuelve su traducción a proteína
    """Traduce un ORF (ADN) a proteína; se detiene en el primer STOP."""
    aa = []                            # lista vacía para ir guardando aminoácidos traducidos
    for i in range(0, len(orf_nt) - 2, 3):  # recorre la secuencia de 3 en 3 (cada codón)
        codon = orf_nt[i:i+3]               # extrae el codón actual (tres bases consecutivas)
        aa1 = CODON.get(codon, "X")         # busca el aminoácido correspondiente en la tabla CODON; "X" si no existe
        if aa1 == "*":                      # si encuentra un STOP, termina la traducción
            break
        aa.append(aa1)                      # añade el aminoácido traducido a la lista
    return "".join(aa)                      # une los aminoácidos en una cadena de texto (proteína)


# para analizar el mejor ORF 
def analizar_mejor_orf(best, sec2):  # recibe el mejor ORF (tupla) y la secuencia original de ADN
    """
    best: tupla (hebra, frame, inicio, fin, orf_nt) devuelta por mejor_orf()
    sec2: secuencia original de ADN (para comparar longitudes)
    """
    if not best:                                # si no existe un ORF válido
        print("No hay ORF para analizar.")      # mensaje de aviso si no hay ORF a analizar
        return None                             # terminamos la función sin devolver nada

    # Desempaquetamos la tupla 'best' en variables con nombres descriptivos
    hebra, frame, start, end, orf_nt = best

    # Traducimos el ORF de ADN a proteína usando la función anterior
    prot = traducir_orf(orf_nt)

    # Contamos cuántos aminoácidos aromáticos (F, Y, W) hay en la proteína
    n_arom = sum(1 for a in prot if a in AROM)

    # Calculamos el porcentaje de aromáticos sobre el total de la proteína
    # Si la proteína está vacía, devolvemos 0 para evitar división por cero
    pct_arom = (100.0 * n_arom / len(prot)) if prot else 0.0

    # Calculamos qué porcentaje del ADN ocupa este ORF respecto a la secuencia completa
    pct_rel_len = 100.0 * len(orf_nt) / len(sec2)

    # Impresión de resultados de manera bonita
    print("\n=== Análisis de la mejor predicción ===")
    print(f"Hebra: {hebra} | Frame: {frame} | Posición: {start}-{end}")
    print(f"Longitud ORF: {len(orf_nt)} nt  |  Longitud proteína: {len(prot)} aa")
    print(f"% aminoácidos aromáticos (F,Y,W): {pct_arom:.2f}%  (n={n_arom})")
    print(f"Longitud relativa al ADN: {pct_rel_len:.2f}%")
    # (opcional) si quieres ver la proteína completa, descomenta la línea siguiente:
    print(f"Proteína:\n{prot}")

    # Devolvemos un diccionario con los resultados numéricos principales
    return {
        "protein": prot,                       # secuencia de la proteína traducida
        "aa_len": len(prot),                   # longitud de la proteína (en aminoácidos)
        "aromatic_percent": pct_arom,          # porcentaje de aminoácidos aromáticos
        "relative_length_percent": pct_rel_len # porcentaje del ORF respecto a la longitud del ADN
    }


# 1. Encontramos todos los ORFs de la secuencia sec2 (usando la función que hicimos antes)
orfs = encontrar_orfs(sec2)

# 2. Escogemos el ORF más largo (la mejor predicción)
best = mejor_orf(orfs)

# 3. Analizamos ese ORF más largo: se traduce y se calcula su composición y longitud relativa
stats = analizar_mejor_orf(best, sec2)

    
```

```python
#Aqui intentaré hacer una función que haga todo
def pipeline_orfs_todo_en_una(
    # Parámetro principal: la secuencia de ADN que queremos analizar,cadena de ADN a analizar (puede pasarse o usar sec2 global(string con A, C, G, T)
    #(puede pasarse o usar sec2 global(string con A, C, G, T)
    secuencia=None,                 
    top_k=6,                         # Número de ORFs más largos que queremos mostrar en el ranking (Top-k)
    imprimir=True,                   # Si es True, imprime resultados detallados en pantalla
    mostrar_proteina=False,          # Si es True, imprime la secuencia de la proteína del mejor ORF
    listar_todos=True,               # Si es True, imprime la lista completa de ORFs (ordenados por longitud)
    devolver_todos=True,             # Si es True, devuelve todos los ORFs en el diccionario de salida
    mostrar_secuencias=False         # Si es True, imprime la secuencia NT y AA de cada ORF (puede ser muy largo)
):
   

    # Aquí comprobamos si he pasado una secuencia directamente.
    # Si no, intentamos usar una variable global llamada 'sec2'.
    if secuencia is None:                      # Si no se pasa una secuencia explícitamente
        try:
            secuencia = sec2                   # intentamos usar una variable global llamada 'sec2'
        except NameError:
            raise ValueError("Pasa una secuencia o define la variable global 'sec2'.")
    # Estandarizamos la secuencia:
    # - upper() convierte todo a MAYÚSCULAS (a veces vienen en minúsculas)
    # - split() elimina espacios y saltos de línea
    # - "".join() vuelve a unir las partes en un solo string continuo
    secuencia = "".join(secuencia.upper().split())  # Quitamos espacios/saltos de línea y convertimos a mayúsculas
   
    #  Tablas y parámetros 
    stop_codones = {"TAA", "TAG", "TGA"}             # Conjunto de codones STOP
    comp = {"A":"T","T":"A","C":"G","G":"C"}         # Tabla de complementos para generar la hebra inversa
    n = len(secuencia)                               # Longitud total de la secuencia de ADN (en nucleótidos)
    orfs = []                                        # Lista donde guardaremos todos los ORFs encontrados
    # Tabla genética estándar ADN → aminoácido (1 letra)
    codon2aa = {
        "TTT":"F","TTC":"F","TTA":"L","TTG":"L","CTT":"L","CTC":"L","CTA":"L","CTG":"L",
        "ATT":"I","ATC":"I","ATA":"I","ATG":"M","GTT":"V","GTC":"V","GTA":"V","GTG":"V",
        "TCT":"S","TCC":"S","TCA":"S","TCG":"S","AGT":"S","AGC":"S",
        "CCT":"P","CCC":"P","CCA":"P","CCG":"P",
        "ACT":"T","ACC":"T","ACA":"T","ACG":"T",
        "GCT":"A","GCC":"A","GCA":"A","GCG":"A",
        "TAT":"Y","TAC":"Y",
        "CAT":"H","CAC":"H","CAA":"Q","CAG":"Q",
        "AAT":"N","AAC":"N","AAA":"K","AAG":"K",
        "GAT":"D","GAC":"D","GAA":"E","GAG":"E",
        "TGT":"C","TGC":"C","TGG":"W",
        "CGT":"R","CGC":"R","CGA":"R","CGG":"R","AGA":"R","AGG":"R",
        "GGT":"G","GGC":"G","GGA":"G","GGG":"G",
        "TAA":"*","TAG":"*","TGA":"*"
    }
    arom = {"F","Y","W"}                           # Conjunto de aminoácidos aromáticos (Fenilalanina, Tirosina, Triptófano)
    
    # En el siguiente, la función recibe como entrada una cadena (string) 'nt'
    # que representa una secuencia de ADN, por ejemplo:
    # nt = "ATGCGT"
    #  por lo tanto la cadena complementaria de ATGCGT sería TACGCA.
    # Además, el ADN es antiparalelo: la hebra complementaria
    # se lee en dirección opuesta (3'→5'). Por eso, además de
    # sustituir cada base por su complemento, debemos invertir
    # el orden de la secuencia (reverse).
    # Aquí usamos una comprensión de lista dentro de un join:
    # 1. nt[::-1] → invierte la secuencia (reverse)
    # 2. comp[b]  → toma el complemento de cada base 'b' según el diccionario 'comp'
    #    (por ejemplo, comp = {"A":"T","T":"A","C":"G","G":"C"})
    # 3. "".join(...) → une todas las bases complementarias en un string único
    def revcomp(nt):
        """Devuelve la cadena complementaria inversa (reverse-complement) de una secuencia de ADN."""
        return "".join(comp[b] for b in nt[::-1])   # Se invierte la cadena y se toma el complemento de cada base

    def traducir_hasta_stop(nt):
        """Traduce una secuencia de ADN en aminoácidos hasta encontrar el primer STOP (excluido)."""
        aa = []                                     # Lista para acumular los aminoácidos
        for i in range(0, len(nt)-2, 3):            # Recorremos de 3 en 3 (codones)
            cod = nt[i:i+3]                         # Tomamos un codón
            aa1 = codon2aa.get(cod, "X")            # Convertimos a aa; "X" si el codón es raro/desconocido
            if aa1 == "*":                          # Si encontramos STOP (*)
                break                               # Terminamos la traducción
            aa.append(aa1)                          # Añadimos el aminoácido
        return "".join(aa)                          # Devolvemos la proteína traducida como string
    # 2) ORFs en hebra positiva 
    for frame in range(3):                          # Se analizan los 3 posibles marcos de lectura: 0, 1 y 2
        i = frame                                   # Posición inicial del frame
        while i <= n - 3:                           # Mientras haya codones completos
            if secuencia[i:i+3] == "ATG":           # Detectamos un codón de inicio (ATG)
                j = i + 3                           # Avanzamos al siguiente codón
                while j <= n - 3:                   # Buscamos STOP dentro del mismo marco
                    if secuencia[j:j+3] in stop_codones:    # STOP encontrado
                        orf_nt = secuencia[i:j+3]           # Secuencia del ORF (incluye STOP final)
                        orfs.append({                       # Guardamos los datos del ORF en un diccionario
                            "hebra": "positive",            # Nombre de la hebra
                            "strand": "+",                  # Símbolo del sentido
                            "frame": frame,                 # Marco de lectura
                            "start": i,                     # Posición inicial (0-based)
                            "end": j+3,                     # Posición final (exclusiva)
                            "len_nt": len(orf_nt),          # Longitud en nucleótidos (incluye STOP)
                            "seq_nt": orf_nt,               # Secuencia de ADN
                            "seq_aa": traducir_hasta_stop(orf_nt)   # Secuencia traducida de aminoácidos (sin STOP)
                        })
                        break                               # Rompemos porque ya encontramos un ORF completo
                    j += 3                                  # Avanzamos 3 bases más (siguiente codón)
            i += 3                                          # Avanzamos 3 bases en la lectura
    # 3) ORFs en hebra negativa 
    for frame in range(3):                          # Igual que antes, pero en la hebra inversa
        i = n - frame - 3                           # Empezamos desde el final según el marco
        while i >= 0:                               # Vamos hacia atrás (de derecha a izquierda)
            codon_rc_view = "".join(comp[b] for b in secuencia[i:i+3][::-1])   # Codón “visto” en la reverse-complement
            if codon_rc_view == "ATG":              # Si vemos un inicio en la RC
                j = i
                while j >= 0:                       # Buscamos STOP hacia atrás
                    stop_rc_view = "".join(comp[b] for b in secuencia[j:j+3][::-1])  # Codón visto como RC
                    if stop_rc_view in stop_codones:  # Si es STOP, tenemos un ORF completo
                        orf_nt_original = secuencia[j:i+3]         # Tramo en coordenadas originales (incluye STOP)
                        orf_nt_rc = revcomp(orf_nt_original)       # Reverse-complement para traducir correctamente
                        orfs.append({
                            "hebra": "negative",                   # Hebra negativa
                            "strand": "-",                         # Indicamos el sentido
                            "frame": frame,                        # Marco de lectura (0,1,2)
                            "start": j,                            # Inicio (0-based)
                            "end": i+3,                            # Fin (exclusivo)
                            "len_nt": len(orf_nt_original),        # Longitud en nt
                            "seq_nt": orf_nt_original,             # Secuencia de ADN tal cual en el genoma
                            "seq_aa": traducir_hasta_stop(orf_nt_rc)  # Traducción de la RC
                        })
                        break
                    j -= 3                             # Retrocedemos 3 bases (un codón)
            i -= 3                                     # Retrocedemos 3 bases en el bucle principal
    #4) Si no se encuentran ORFs
    if not orfs:
        if imprimir:
            print("No se encontraron ORFs.")           # Mensaje informativo explicando que no se encontraron ORF
        salida_vacia = {                               # Diccionario vacío coherente
            "orfs_totales": 0,
            "top_orfs": [],
            "mejor_orf": None,
            "proteina": "",
            "aa_len": 0,
            "porcentaje_aromaticos": 0.0,
            "longitud_relativa": 0.0
        }
        if devolver_todos:
            salida_vacia["todos_los_orfs"] = []        # Añadimos lista vacía
        return salida_vacia
    #  5) Ordenar ORFs por longitud 
    # Ordena la lista de ORFs (diccionarios) por la longitud en nucleótidos ('len_nt')
    #'reverse=True' → orden descendente (del más largo al más corto)
    orfs_ordenados = sorted(orfs, key=lambda d: d["len_nt"], reverse=True)  # Orden descendente
    top_orfs = orfs_ordenados[:top_k]               # Seleccionamos los top-k (por defecto top_k=6)
    best = orfs_ordenados[0]                        # El más largo es el "mejor"
    # 6) Calcular métricas del mejor ORF
    proteina = best["seq_aa"]                       # Secuencia de proteína (sin STOP)
    aa_len = len(proteina)                          # Longitud de la proteina en aminoácidos
    n_arom = sum(1 for a in proteina if a in arom)  # Contamos F/Y/W
    pct_arom = (100.0 * n_arom / aa_len) if aa_len else 0.0   # % de aromáticos
    pct_rel_len = 100.0 * best["len_nt"] / n        # % longitud del ORF frente al ADN total

    
    # 7) IMPRESIÓN EN CONSOLA
    # Este bloque solo se ejecuta si el pido ver los resultados
    # por pantalla (imprimir=True). Muestra:
    #   (a) La lista completa de ORFs, ordenados por longitud.
    #   (b) El TOP-K de ORFs más largos.
    #   (c) Un resumen detallado del ORF "best" (el más largo).
    # Además, opcionalmente imprime las secuencias NT/AA completas
    # si mostrar_secuencias=True y la proteína si mostrar_proteina=True.

    if imprimir:  

    # (a) TODOS los ORFs ordenados por longitud (de mayor a menor)
    #     Solo si listar_todos=True para no saturar la salida.
     if listar_todos:
        print("\n=== TODOS los ORFs (ordenados por longitud) ===")

        # Recorremos la lista 'orfs_ordenados' ya ordenada previamente.
        # enumerate(..., 1) numera las filas empezando en 1 (más humano).
        for i, d in enumerate(orfs_ordenados, 1):

            # Cada 'd' es un diccionario con los metadatos de un ORF:
            #   d['strand']  : '+' o '-' según hebra directa/complementaria
            #   d['hebra']   : etiqueta descriptiva de la hebra (p.ej., 'directa'/'inversa')
            #   d['frame']   : marco de lectura (0, 1 o 2)
            #   d['start']   : posición inicial del ORF (en nucleótidos)
            #   d['end']     : posición final del ORF (en nucleótidos)
            #   d['len_nt']  : longitud del ORF en nucleótidos
            #   d['seq_aa']  : secuencia traducida a aminoácidos (sin STOP final)
            # Mostramos una línea resumen compacta para cada ORF.
            print(
                f"{i:>3}. "                                  # índice con ancho 3, alineado a la derecha
                f"Strand: {d['strand']} ({d['hebra']}) | "   # hebra + etiqueta
                f"Frame: {d['frame']} | "                    # marco de lectura
                f"Pos: {d['start']}-{d['end']} | "           # rango de posiciones
                f"Long: {d['len_nt']} nt | "                 # longitud en nucleótidos
                f"AA: {len(d['seq_aa'])}"                    # longitud en aminoácidos
            )

            # Si el quiero ver las secuencias completas de cada ORF:
            if mostrar_secuencias:
                #   d['seq_nt'] : secuencia de nucleótidos del ORF (desde start hasta end)
                #   d['seq_aa'] : secuencia de aminoácidos traducida (sin STOP)
                print("    NT:", d["seq_nt"])
                print("    AA:", d["seq_aa"])
                print("-" * 70)  # línea separadora para legibilidad

   
    # (b) TOP-K: resumen solo de los ORFs más largos (más conciso)
    print("\n=== Los ORFs más largos (top-k) ===")

    # 'top_orfs' es el corte orfs_ordenados[:top_k] hecho antes.
    for i, d in enumerate(top_orfs, 1):
        print(
            f"{i}. "                                   # índice del 1 al K
            f"Strand: {d['strand']} ({d['hebra']}) | " # hebra
            f"Frame: {d['frame']} | "                  # marco
            f"Pos: {d['start']}-{d['end']} | "         # posiciones
            f"Long: {d['len_nt']} nt"                  # longitud en nt
        )

   
    # (c) MEJOR PREDICCIÓN: el ORF más largo ('best')
    #     Incluye métricas adicionales calculadas antes:
    #       - aa_len: longitud de la proteína (en aminoácidos)
    #       - n_arom: nº de aa aromáticos (F, Y, W)
    #       - pct_arom: % de aa aromáticos
    #       - pct_rel_len: % de longitud del ORF respecto al ADN total
   
    print("\n=== Mejor predicción (ORF más largo) ===")
    print(
        f"Strand: {best['strand']} ({best['hebra']}) | "
        f"Frame: {best['frame']} | "
        f"Pos: {best['start']}-{best['end']} | "
        f"Long nt: {best['len_nt']}"
    )

    # Línea de métricas bioquímicas y de cobertura genómica
    print(
        f"Proteína (aa): {aa_len} | "
        f"% aromáticos (F,Y,W): {pct_arom:.2f}% (n={n_arom}) | "
        f"% ORF vs ADN: {pct_rel_len:.2f}%"
    )

    # Si el usuario quiere ver la SECUENCIA de proteína del mejor ORF:
    if mostrar_proteina:
        print("\nSecuencia de proteína (STOP excluido):")
        print(proteina)  # 'proteina' = best['seq_aa'] calculado previamente

          

resultados = pipeline_orfs_todo_en_una(
    sec2,
    top_k=6,
    imprimir=True,
    mostrar_proteina=False,
    listar_todos=True,     #  esto hace que imprima TODOS los ORFs encontrados
    devolver_todos=True    # esto hace que también los devuelva en el resultado
)
```
