#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
def digerir():
    cortes_por_enz = {}
    sin_sitio = []
    todos_cortes = []

    for nombre, patron in enzR.items():
        star_pos = patron.index("*")
        motivo = patron.replace("*","")
        posiciones = []
        i = 0
        while True:
            idx = sec2.find(motivo, i)
            if idx == -1:
                break
            corte0 = idx + star_pos
            posiciones.append(corte0+1)
            todos_cortes.append(corte0)
            i = idx + 1
        if posiciones:
            cortes_por_enz[nombre] = posiciones
        else:
            sin_sitio.append(nombre)

    # Construir fragmentos
    todos_cortes = sorted(set(todos_cortes))
    fragmentos = []
    prev = 0
    for c in todos_cortes:
        fragmentos.append(sec2[prev:c])
        prev = c
    fragmentos.append(sec2[prev:])
    tamaños = [len(f) for f in fragmentos]

    return cortes_por_enz, sin_sitio, fragmentos, tamaños

# --- Uso ---
cortes, no_sites, frags, sizes = digerir()
print(cortes)
print(no_sites)
print(frags)
print(sizes)
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# --- Tabla de traducción estándar (stop = '*') ---
codon2aa = {
    'TTT':'F','TTC':'F','TTA':'L','TTG':'L','TCT':'S','TCC':'S','TCA':'S','TCG':'S',
    'TAT':'Y','TAC':'Y','TAA':'*','TAG':'*','TGT':'C','TGC':'C','TGA':'*','TGG':'W',
    'CTT':'L','CTC':'L','CTA':'L','CTG':'L','CCT':'P','CCC':'P','CCA':'P','CCG':'P',
    'CAT':'H','CAC':'H','CAA':'Q','CAG':'Q','CGT':'R','CGC':'R','CGA':'R','CGG':'R',
    'ATT':'I','ATC':'I','ATA':'I','ATG':'M','ACT':'T','ACC':'T','ACA':'T','ACG':'T',
    'AAT':'N','AAC':'N','AAA':'K','AAG':'K','AGT':'S','AGC':'S','AGA':'R','AGG':'R',
    'GTT':'V','GTC':'V','GTA':'V','GTG':'V','GCT':'A','GCC':'A','GCA':'A','GCG':'A',
    'GAT':'D','GAC':'D','GAA':'E','GAG':'E','GGT':'G','GGC':'G','GGA':'G','GGG':'G'
}

def revcomp(dna):
    mapa = str.maketrans("ACGTacgt", "TGCAtgca")
    return dna.translate(mapa)[::-1]

def traducir(dna):
    aa = []
    for i in range(0, len(dna)-2, 3):
        aa.append(codon2aa.get(dna[i:i+3], 'X'))
    return "".join(aa)

def encontrar_orfs(dna):
    """Devuelve lista de ORFs como diccionarios con info completa en + y -."""
    s = dna.upper().replace("\n","").replace(" ","")
    L = len(s)
    r = revcomp(s)

    orfs = []

    # ---- Strand positiva (+) ----
    for frame in range(3):
        i = frame
        while i <= L-3:
            codon = s[i:i+3]
            if codon == "ATG":
                j = i + 3
                while j <= L-3 and s[j:j+3] not in ("TAA","TAG","TGA"):
                    j += 3
                if j <= L-3 and s[j:j+3] in ("TAA","TAG","TGA"):
                    nt_seq = s[i:j+3]
                    aa_seq = traducir(nt_seq)[:-1]  # sin el stop final
                    orfs.append({
                        "strand": "+",
                        "frame": frame,
                        "start_nt_1based": i+1,
                        "end_nt_1based": j+3,
                        "len_nt": j+3 - i,
                        "len_aa": len(aa_seq),
                        "aa": aa_seq
                    })
                    i = j + 3
                    continue
            i += 3

    # ---- Strand negativa (-) (buscar en revcomp y mapear coordenadas) ----
    for frame in range(3):
        i = frame
        while i <= L-3:
            codon = r[i:i+3]
            if codon == "ATG":
                j = i + 3
                while j <= L-3 and r[j:j+3] not in ("TAA","TAG","TGA"):
                    j += 3
                if j <= L-3 and r[j:j+3] in ("TAA","TAG","TGA"):
                    nt_seq_rc = r[i:j+3]
                    aa_seq = traducir(nt_seq_rc)[:-1]
                    # mapear a coords del original:
                    # revcomp[i:j+3] corresponde a original[ L-(j+3) : L-i ]
                    start0 = L - (j+3)
                    end0   = L - i
                    orfs.append({
                        "strand": "-",
                        "frame": frame,
                        "start_nt_1based": start0+1,
                        "end_nt_1based": end0,
                        "len_nt": (end0 - start0),
                        "len_aa": len(aa_seq),
                        "aa": aa_seq
                    })
                    i = j + 3
                    continue
            i += 3

    return orfs

def analisis_orfs(dna):
    """Hace todo lo pedido y muestra un resumen claro."""
    s = dna.upper().replace("\n","").replace(" ","")
    L = len(s)
    orfs = encontrar_orfs(s)

    # 1) Todos los ORFs (ya los tenemos en `orfs`)
    total_orfs = len(orfs)

    # 2) Los 6 más largos
    top6 = sorted(orfs, key=lambda d: d["len_aa"], reverse=True)[:6]

    # 3) Mejor predicción = el más largo
    best = top6[0] if top6 else None

    # 4) % aromáticos y longitud relativa del mejor
    if best and best["len_aa"] > 0:
        aa = best["aa"]
        arom = aa.count("F") + aa.count("Y") + aa.count("W")  # F, Y, W
        pct_aromaticos = arom / len(aa) * 100
        pct_longitud_rel = best["len_nt"] / L * 100
    else:
        pct_aromaticos = 0.0
        pct_longitud_rel = 0.0

    # ---------- Salida legible ----------
    print("=== ORF ANALYSIS ===")
    print(f"Longitud ADN: {L} nt")
    print(f"ORFs encontrados: {total_orfs}")

    if total_orfs == 0:
        print("No se encontraron ORFs.")
        return orfs, top6, best, pct_aromaticos, pct_longitud_rel

    print("\nTop 6 ORFs más largos (len_aa, strand, start..end nt):")
    for k, d in enumerate(top6, 1):
        print(f"  {k}. {d['len_aa']} aa | strand {d['strand']} | {d['start_nt_1based']}..{d['end_nt_1based']}")

    print("\nMejor predicción (más largo):")
    print(f"  {best['len_aa']} aa | strand {best['strand']} | {best['start_nt_1based']}..{best['end_nt_1based']}")
    print(f"  % aromáticos (F,Y,W): {pct_aromaticos:.2f}%")
    print(f"  Longitud relativa al ADN: {pct_longitud_rel:.2f}%")
    # (Opcional) ver primeros/últimos aa
    if best['len_aa'] > 30:
        print(f"  aa: {best['aa'][:15]}...{best['aa'][-15:]}")

    # Devolver por si necesitas usarlo después
    return orfs, top6, best, pct_aromaticos, pct_longitud_rel
```
